<h1> How to calculate relative permeability of oil and water</h1>

<div class='alert alert block alert-info'> relative permeability = (effective permeability)/(absolute permeability)</div>

absolute permeability = permeability at 100% saturation of that fluid

for ex. ` if only two phase are present then at water saturation (sw) = 0 , oil saturation will be (so) = 100 so its gives us absolute permeability of oil`

```python
kro = df.qo/df.qo[0]  #q at per qo(sw = 0)
krw = df.qw/df.qw[9]

plt.plot(df.Sw, kro, label = 'oil relative permeability (kro)')
plt.plot(df.Sw, krw, label = 'water relative permeability (krw)')
plt.xlabel('Sw')
plt.ylabel('kro and krw')
plt.title('Relative Permeability')
plt.legend()
```

<h1> How to calculate phase permeabilities of oil and water </h1>

> We use Darcy equation for this:
    <div class='alert alert block alert-info'> **q= -(k*A*dp)/(mu*dx)** 
    
        Here negative sign because of pressure drop
        unit:
               q = flow rate, cm3/sec
               k = Permeability ,D
               A = area, cm2
               dp = Pressure drop, atm
               mu = viscosity, cp
               dx = lenght, cm </div>
               
               
               
  > Given:
            mu_oil = 5 cp
            mu_water = 1.2 cp
            core_lenght = 12 cm
            dp = -2.5 atm
            core_area = 5 cm2
    

```python
mu_oil = 5
mu_water = 1.2
core_lenght = 12
dp = -2.5
core_area = 5

ko = -(df.qo*mu_oil*core_lenght)/(dp*core_area)
kw = -(df.qw*mu_water*core_lenght)/(dp*core_area)

# plotting graph between 
plt.plot(df.Sw, kro, label = 'oil permeability (ko)')
plt.plot(df.Sw, krw, label = 'water permeability (kw)')
plt.xlabel('Sw')
plt.ylabel('ko and kw')
plt.title('Phase Permeability')
plt.legend()
```
